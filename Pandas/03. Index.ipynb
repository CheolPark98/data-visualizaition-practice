# 3. Index
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
DSC = {
    '이름' : ['렌고쿠', '무이치로', '텐겐', '미츠리', '기유', '시노부', '사네미', '탄지로'],
    '계급' : ['주', '주', '주', '주', '주', '주', '주', '병'],
    '키' : [177, 160, 198, 167, 176, 151, 179, 165],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    '사용호흡' : ['화염', '안개', '소리', '사랑', '물', '벌레', '바람', '물, 해']
}
DSC
```

```python
df = pd.DataFrame(DSC, index=['염주','하주','음주','연주','수주','충주','풍주','주인공'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name ='이명'
df
```

## Index 초기화

```python
df.reset_index() #index를 column으로 편집
```

```python
df.reset_index(drop=True) # 원래 쓰던 인덱스를 삭제하면 어떻게 되는지 보여줌
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

## Index 설정
지정한 column으로 Index설정

```python
df.set_index('이름')
df
```

## Index 정렬

```python
df.sort_index() #인덱스로 오름차순 정렬
```
