```python
import pandas as pd
df = pd.read_csv('DSC.txt',index_col='이명' )
df
```

```python
df
```

```python
df['사용호흡'] = df['사용호흡']+'의 호흡'
df
```

```python
df.loc[df['사용호흡']=='물, 해의 호흡','사용호흡']='시작의 호흡'
df
```
