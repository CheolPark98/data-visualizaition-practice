# 11. 데이터 정렬

```python
import pandas as pd
df = pd.read_csv('DSC.txt',index_col='이명' )
df  
```

```python
df.sort_values('키') #오름차순으로 정렬
```

```python
df.sort_values('키',ascending=False)
```

```python
df.sort_values(['수학','영어'],ascending=False)
```

```python
df.sort_values(['수학','영어'],ascending=[False,True])
```
