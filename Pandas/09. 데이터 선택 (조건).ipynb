# 9. 데이터 선택 (조건)

```python
import pandas as pd
df = pd.read_csv('DSC.txt',index_col='이명' )
df  
```

```python
df['키'] >= 180
```

```python
fit=(df['키']>=180) # 필터
df[fit]
```

```python
df[~fit]  # 필터를 역으로 적용
```

```python
df.loc[df['키']<=185,'키']
```

```python
df.loc[(df['키']<=185 )& (df['사용호흡'] =='물')]
```

### | 또는

```python
df.loc[(df['키']>=185)|(df['키']<=170)]
```

### str 함수

```python
filt = df['이름'].str.startswith('무')
df[filt]
```

```python
filt=df['이름'].str.contains('츠')
df[filt]
```

```python
df[~filt]
```

```python
breas=['안개','소리']
filt=df['사용호흡'].isin(breas)
df[filt]
```
