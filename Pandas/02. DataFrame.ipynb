# DataFrame
2차원 데이터 (Series 들의 모음)

## Data 준비

사전 (dict) 자료구조를 통해 생성

```python
data = {
    '이름' : ['렌고쿠', '무이치로', '텐겐', '미츠리', '기유', '시노부', '사네미', '탄지로'],
    '계급' : ['주', '주', '주', '주', '주', '주', '주', '병'],
    '키' : [177, 160, 198, 167, 176, 151, 179, 165],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    '사용호흡' : ['화염', '안개', '소리', '꽃', '물', '벌레', '바람', '물, 해']
}
data
```

## DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

## 데이터 접근

```python
df["이름"]
```

```python
df[['이름','사용호흡']] #두개 이상의 데이터를 가져오려면 리스트로 감싸야함
```

 ## DataFrame 객체 생성 (Index 지정)

```python
df = pd.DataFrame(data, index=['염주','하주','음주','연주','수주','충주','풍주','주인공'])
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column만 선택하거나 순서 변경 가능

```python
dd = pd.DataFrame(data,columns=['이름','사용호흡'])
dd
```
